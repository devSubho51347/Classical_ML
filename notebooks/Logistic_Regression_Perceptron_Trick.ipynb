```python
### Install the libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_classification
```

```python
### Make classification dataset

# Generate a simple binary classification dataset
X, y = make_classification(
    n_samples=500,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42,
    class_sep = 1.5,
    n_clusters_per_class = 2
    
)

# Plot the dataset
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Basic Binary Classification Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Class')
plt.show()
```

# Perceptron Trick

```python
# ### We have inserted 1 for the bias term that we add to all the data points

# XX = np.insert(X,0,1, axis = 1)
# XX
```

```python
### define the perceptron method

# define step

def step(x):
    return 1 if x > 0 else 0

def perceptron(X,y):

    XX = np.insert(X,0,1,axis = 1)
    ##initialize the weights
    weights = np.ones(XX.shape[1])
    lr = 0.01
    #### Train the model for 100 epochs
    for i in range(100):
        j = np.random.randint(0,200)
        yhat = step(np.dot(XX[j],weights))
        weights = weights + lr*(y[j]-yhat)*XX[j]
    return weights[0],weights[1:]



    

```

```python
intercept, coeff = perceptron(X,y)
```

```python
intercept, coeff
```

```python
m = coeff[0]/coeff[1]
b = intercept/coeff[1]
m,b

x_input = np.linspace(-5,5,100)
y_output = m*x_input + b
```

```python
y_output
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input,y_output,color = 'black', linewidth = 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Basic Binary Classification Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Class')
plt.show()
```

# Perceptron Trick Using Sigmoid

```python
def step_sigmoid(z):
    y = 1/(1 + np.exp(-z))
    return y

def perceptron(X,y,lr):
    X = np.insert(X,0,1,axis = 1)
    lr = lr
    weights = np.ones(X.shape[1])

    for j in range(10):
        for i in range(200):
            k = np.random.randint(0,200)
            yhat = step_sigmoid(np.dot(X[k],weights))
            weights = weights + lr*(y[k] - yhat)*X[k]
            
    return weights[0],weights[1:]
    
    
```

```python
intercept, coeff = perceptron(X,y,0.01)
```

```python
intercept, coeff
```

```python
m = coeff[0]/coeff[1]
b = intercept/coeff[1]
m,b

x_input = np.linspace(-0.5,0.5,100)
y_output = m*x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input,y_output,color = 'black', linewidth = 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Basic Binary Classification Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Class')
plt.show()
```
